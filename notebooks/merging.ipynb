### Loading Datasets

```python
import pandas as pd

enrol = pd.read_csv(
    "Aadhar Enrollment Dataset for Thane.csv",
    parse_dates=['date']
)

bio = pd.read_csv(
    "Aadhaar Biometric Monthly Update Data.csv",
    parse_dates=['date']
)

demo = pd.read_csv(
    "Aadhar Demographic Updates for Thane.csv",
    parse_dates=['date']
)
```

```python
enrol = enrol[[
    'date', 'state', 'district', 'pincode',
    'age_0_5', 'age_5_17', 'age_18_greater'
]]
```

```python
demo = demo[[
    'date', 'state', 'district', 'pincode',
    'demo_age_5_17', 'demo_age_17_'
]]
```

```python
bio = bio[[
    'date', 'state', 'district', 'pincode',
    'bio_age_5_17', 'bio_age_17_'
]]
```

### Merging the columns 

```python
merged = pd.merge(
    enrol,
    demo,
    on=['date', 'state', 'district', 'pincode'],
    how='outer'
)
```

```python
merged.head()
```

```python
merged.shape
```

```python
merged = pd.merge(
    merged,
    bio,
    on=['date', 'state', 'district', 'pincode'],
    how='outer'
)
```

```python
merged.head()
```

```python
merged.shape
```

```python
merged.isnull().sum()
```

### Handling NaN values

```python
numeric_cols = [
    'age_0_5', 'age_5_17', 'age_18_greater',
    'demo_age_5_17', 'demo_age_17_',
    'bio_age_5_17', 'bio_age_17_'
]

merged[numeric_cols] = merged[numeric_cols].fillna(0)
```

### Sanity check for merged columns

```python
merged.isna().sum()
```

```python
merged[numeric_cols].dtypes
```

```python
print("Enrol rows:", enrol.shape[0])
print("Demo rows:", demo.shape[0])
print("Bio rows:", bio.shape[0])
print("Merged rows:", merged.shape[0])
```

### Save the merged file 

```python
merged = merged.sort_values(['pincode', 'date'])

merged.to_csv(
    'D:/UIDAI HACKATHON/UIDAI-Hackathon/data/processed/Aadhaar_daily.csv',
    index=False
)
```

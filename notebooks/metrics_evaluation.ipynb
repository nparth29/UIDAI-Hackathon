# ICI (Identity Churn Index)
### ICI answers this question:
###  In this PIN, in this month, how much of Aadhaar activity is driven by adults repeatedly changing their identity details?

### Formula:
### 1) adult_changes (i.e total demographic and biometric updates made by adults)
### 2) total_transactions (all enrolments + all updates in that pincode)
### Hence ICI = adult_changes / total_transactions

### Conclusion:
###  Voluntary updates = stress / migration / instability

```python
# computing new enrolments per month,  per PIN

m['new_enrol'] = (
    m['age_0_5'] +
    m['age_5_17'] +
    m['age_18_greater']
)
```

```python
m.head(20)
```

```python
# computing total transactions per month, per PIN
m['total_txn'] = (
    m['new_enrol'] +
    m['demo_age_5_17'] +
    m['demo_age_17_'] +
    m['bio_age_5_17'] +
    m['bio_age_17_']
)
```

```python
m.head(20)
```

```python
# computing ICI 
# 1 is put here to avoid denominator being 0 

m['ICI'] = (m['demo_age_17_'] + m['bio_age_17_']) / (m['total_txn'] + 1)
```

```python
# creating threshold values 

ici_90 = m['ICI'].quantile(0.90)
ici_75 = m['ICI'].quantile(0.75)

ici_90, ici_75
```

```python
# Flag PIN-MONTHS
m['ici_top_10pct'] = m['ICI'] >= ici_90
m['ici_top_25pct'] = m['ICI'] >= ici_75
```

### Aggregating PIN codes which have higher ICI more frequently is much better than just providing which PIN codes have a higher ICI value.

```python
# count how often each PIN appears in top 10 percent 

pin_ici_freq = (
    m.groupby('pincode')
     .agg(
         months_total=('month', 'nunique'),
         top10_count=('ici_top_10pct', 'sum'),
         top25_count=('ici_top_25pct', 'sum')
     )
     .reset_index()
)

pin_ici_freq.head(30)
```

```python
# Counts alone are misleading if PINs have different data lengths. Hence we compute frequency

pin_ici_freq['top10_frac'] = (
    pin_ici_freq['top10_count'] / pin_ici_freq['months_total']
)

pin_ici_freq['top25_frac'] = (
    pin_ici_freq['top25_count'] / pin_ici_freq['months_total']
)
```

```python
# Top persistent extreme churn PINs

pin_ici_freq.sort_values(
    'top10_frac',
    ascending=False
).head(10)
```

# BDR (Biometric Distress Ratio)

### BDR answers this question

### When adults update Aadhaar in this PIN and month, how much of that updating is driven by biometric problems rather than normal demographic changes?

### Why BDR complements ICI (important)

### High ICI + High BDR → biometric distress zone

### High ICI + Low BDR → migration / address churn

### Low ICI + High BDR → isolated biometric problems

# Formula:

### 1) adult biometric updates (total biometric updates made by adults)
### 2) adult demographic updates (total demographic updates made by adults)

### Hence,
### BDR = adult biometric updates / (adult demographic updates + 1)

# Conclusion:

### So BDR tells us: Is biometric failure dominating adult updates here?

```python
# computing BDR 
m['BDR'] = m['bio_age_17_'] / (m['demo_age_17_'] + 1)
```

```python
m.head()
```

```python
# Sanity Check

m['BDR'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
```

```python

m[['pincode','month','bio_age_17_','demo_age_17_','BDR']].head(10)
```

```python
# FLAG the rows where there are 0 demographic updates and only biometric updates 
m['biometric_only'] = m['demo_age_17_'] == 0
```

```python
m.head()
```

```python
# computing BDR only for the rows having some demographic updates as well and keeping empty values in rows having no demographic updates 

m['BDR_valid'] = None
mask = m['demo_age_17_'] > 0
m.loc[mask, 'BDR_valid'] = m.loc[mask, 'bio_age_17_'] / m.loc[mask, 'demo_age_17_']
```

```python
m.head()
```

```python
cap = m['BDR_valid'].quantile(0.95)
cap
```

```python
m['BDR_capped'] = m['BDR_valid'].clip(upper=cap)
```

```python
m.head()
```

```python
# tells us how many months each pin had where there were only and only biometric updates
bd_pin = (
    m.groupby('pincode')
     .agg(
         months_total=('month', 'nunique'),
         biometric_only_count=('biometric_only', 'sum')
     )
     .reset_index()
)

bd_pin.head()
```

```python
# convert into frequency or fraction due to data length inconcisitency

bd_pin['biometric_only_fraction'] = (
    bd_pin['biometric_only_count'] / bd_pin['months_total']
)
```

```python
bd_pin.head(20)
```

```python
# tells when demographic updates DID happen, how strong was biometric pressure

bd_stats = (
    m.groupby('pincode')
     .agg(
         median_BDR=('BDR_capped', 'median'),
         mean_BDR=('BDR_capped', 'mean')
     )
     .reset_index()
)

bd_stats.head(20)
```

```python
# combining the "biometrics_only" data with the normal bdr values

bd_pin = bd_pin.merge(bd_stats, on='pincode', how='left')
```

```python
bd_pin.head(20)
```

```python
bd_pin.describe()
```

```python
print(bd_pin['biometric_only_fraction'].quantile(0.75))
print(bd_pin['median_BDR'].quantile(0.75))
```

```python
# finding the pins which are much worse than others 

bd_pin['high_biometric_only'] = (
    bd_pin['biometric_only_fraction'] >=
    bd_pin['biometric_only_fraction'].quantile(0.75)
)

bd_pin['high_BDR'] = (
    bd_pin['median_BDR'] >=
    bd_pin['median_BDR'].quantile(0.75)
)
```

```python
bd_pin.head(20)
```

```python
# combining the pin-month data together 

pin_summary = pin_ici_freq.merge(
    bd_pin,
    on='pincode',
    how='left'
)
```

```python
pin_summary.head(30)
```

```python
pin_summary.shape
```

# CBCR (Child Biometric Compliance Ratio)

### What CBCR answers (in plain words):
### When children come for Aadhaar updates in a PIN and month, are they actually completing the mandatory biometric update?

# Formula:

### 1) child biometric updates (total biometric updates of children)
### 2) child demographich updates (total demographic updates of children)

### Hence,
### CBCR = child biometric updates / (child biometric updates + child demographic updates + 1)

```python
# computing the CBCR 

m['CBCR'] = (
    m['bio_age_5_17'] /
    (m['bio_age_5_17'] + m['demo_age_5_17'] + 1)
)
```

```python
m.head(50)
```

```python
m['CBCR'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```

```python
# summarize how each PIN behaves overall.
cbcr_pin = (
    m.groupby('pincode')
     .agg(
         months_total=('month', 'nunique'),
         mean_CBCR=('CBCR', 'mean'),
         median_CBCR=('CBCR', 'median')
     )
     .reset_index()
)

cbcr_pin.head(50)
```

```python
cbcr_cutoff = cbcr_pin['mean_CBCR'].quantile(0.25)
cbcr_cutoff
```

```python
cbcr_pin['low_CBCR'] = cbcr_pin['mean_CBCR'] <= cbcr_cutoff
```

```python
cbcr_pin.head()
```

```python
# merging into pin_summary
pin_summary = pin_summary.merge(
    cbcr_pin[['pincode', 'mean_CBCR', 'median_CBCR', 'low_CBCR']],
    on='pincode',
    how='left'
)
```

```python
pin_summary.shape
```

# Enrollment Saturation

### What this metric answers 
### In this PIN and month, how much Aadhaar activity is still about creating new Aadhaar numbers versus just maintaining existing ones?

### Low value → area is already enrolled (saturated)
### High value → area still needs enrolment capacity

# Formula:
### 1) New_enrol_pin (all the enrolments in a particular month)
### 2) Total_txn_pin (all the transactions in that month which include new_enrol_pin, all biometric updates and all demographic updates)
### ENROLMENT SATURATION = NEW_ENROL_PIN / (TOTAL_TXN_PIN + 1)


```python
m.columns
```

```python
# computing total aadhar transactions 

m['total_updates'] = (
    m['demo_age_5_17'] +
    m['demo_age_17_'] +
    m['bio_age_5_17'] +
    m['bio_age_17_']
)

m['total_txn'] = m['new_enrol'] + m['total_updates']
```

```python
# computing the enrolment saturation per month 
m['enrol_sat'] = m['new_enrol'] / (m['total_txn'] + 1)
```

```python
m.drop(columns="ENROL_SAT",inplace=True)
```

```python
m
```

```python
m['enrol_sat'].describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9])
```

```python
# calculate the overall pin behaviour for enrolment 

enrol_pin = (
    m.groupby('pincode')
     .agg(
         months_total=('month', 'nunique'),
         mean_ENROL_SAT=('enrol_sat', 'mean'),
         median_ENROL_SAT=('enrol_sat', 'median'),
         mean_new_enrol=('new_enrol', 'mean'),
         mean_total_updates=('total_updates', 'mean')
     )
     .reset_index()
)
```

```python
enrol_pin.head()
```

```python
sat_cutoff = enrol_pin['median_ENROL_SAT'].quantile(0.25)
sat_cutoff
```

```python
enrol_pin['is_saturated'] = enrol_pin['median_ENROL_SAT'] <= sat_cutoff
```

```python
enrol_pin.head(50)
```

```python
pin_summary = pin_summary.merge(
    enrol_pin[['pincode', 'mean_ENROL_SAT', 'median_ENROL_SAT', 'is_saturated']],
    on='pincode',
    how='left'
)
```

```python
pin_summary.head()
```
